```python
import pandas as pd
import numpy as np
#import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.ensemble import AdaBoostClassifier
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score
from sklearn.decomposition import PCA
```

```python
dataset = pd.read_csv('train.csv')
```

```python
dataset2=pd.read_csv('unique_m.csv')
```

```python
dataset_completo=pd.merge(dataset,dataset2)
```

```python
dataset_completo
```

```python
dt=dataset_completo.drop('material',axis=1)
dt
```

```python
y=dt['critical_temp']
y
```

```python
X=dt.drop("critical_temp",axis=1)
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 5)
```

```python
scaler = StandardScaler().fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
n_comps=80
W, v = np.linalg.eig(X_train.T @ X_train)
vexp = W/W.sum()

tve=0
for i, ve in enumerate(vexp[:n_comps]):
    tve+=ve
    print("PC%d - Variance explained: %7.4f - Total Variance: %7.4f" % (i, ve, tve) )
    
```

```python
pca = PCA(n_components=85)
X_train_pca = pca.fit_transform(X_train)
model = DecisionTreeRegressor()
model.fit(X_train_pca, y_train)
scores = cross_val_score(model, X, y, cv=5)
scores
```
